```python
#Arreglos Multidimensionales
arr = np.array([1,2,3,4,5,6], ndmin = 3)

print(arr)
print(arr[0])
print(arr[0][0])
print(arr[0][0][0])
```

#Operaciones

```python
arr1 = [1,2,3]
arr2 = [4,5,6]
arr3 = np.add(arr1, arr2)
print(arr3)
```

```python
#Agregacion
arr = [1,5,7]
media = np.mean(arr)
print(media)
```

```python
#Agregacion
arr = [1,5,7]
maximo = np.max(arr)
print(maximo)
```

```python
#Agregacion
arr = [15,5,7]
minimo = np.min(arr)
print(minimo)
```

```python
#indexacion

arr = np.array([1,2,3,4,5,6])
print(arr[0])
```

```python
#indexacion

arr = np.array([1,2,3,4,5,6])
print(arr[-1])
```

```python
#indexacion

arr = np.array(([1,2,3],[4,5,6]))
print(arr[1,2])
```

```python
#Slicing

arr = np.array([1,2,3,4,5,6])
print(arr[1:4])
```

```python
#Slicing

arr = np.array([1,2,3,4,5,6])
print(arr[1:4:2])
```

```python
#Slicing

arr = np.array([1,2,3,4,5,6])
print(arr[::2])
```
